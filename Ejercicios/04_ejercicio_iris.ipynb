```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Cargar dataset de iris
iris = sns.load_dataset('iris')
df = pd.DataFrame(iris)
df.head()
```

```python
# Mostrar ultimos 5 registros
df.tail()
```

```python
# Mostrar información del dataset
df.info()
```

```python
# Mostrar descripción del dataset
df.describe()
```

```python
# Cantidad de valores nulos
df.isnull().sum()
```

```python
# Gráfico de distribución de la longitud del sépalo
df['sepal_length'].hist()
plt.title('Distribución de la longitud del sépalo')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Grafico de la longitud del pétalo
df['petal_length'].hist(color='red')
plt.title('Distribución de la longitud del pétalo')
plt.xlabel('Longitud del pétalo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de dispersión de la longitud y ancho del sépalo
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal_length'], df['sepal_width'], color='blue')
plt.title('Dispersión de la longitud y ancho del sépalo')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Ancho del sépalo')
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Comparación de longitud del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del pétalo')
plt.show()
```
