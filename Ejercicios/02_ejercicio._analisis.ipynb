```python
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar los datos desde un archivo CSV
df = pd.read_csv('./datos/ventas_mensuales.csv')
df.head()
```

```python
# Descripci√≥n de los datos
df.describe()
```

```python
# Contar los valores nulos
df.isnull().sum()
```

```python
# Graficar los datos
plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['Ventas'], marker='o', color='b', linestyle='-', linewidth=2)
plt.title('Ventas mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid()
plt.show()
```

```python
media_ventas = df['Ventas'].mean()
print(f'Media de ventas: {media_ventas:.2f}')
```

```python
# Mes con m√°s ventas
mes_mas_ventas = df[df['Ventas'] == df['Ventas'].max()]['Mes'].values[0]
print(f'Mes con m√°s ventas: {mes_mas_ventas}')
```

```python
# Gr√°fico de barra de las ventas mensuales
plt.figure(figsize=(10, 6))
plt.bar(df['Mes'], df['Ventas'], color='g')
plt.title('Ventas mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid()
plt.show()
```
